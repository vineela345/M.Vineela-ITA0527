```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def draw_circle_on_image(image, radius, center_x, center_y, color=(0, 0, 0), thickness=2):
    """
    Draws a circle on a given image.

    Args:
        image (numpy.ndarray): The input image.
        radius (int): The radius of the circle.
        center_x (int): The x-coordinate of the circle's center.
        center_y (int): The y-coordinate of the circle's center.
        color (tuple): The BGR color of the circle (default is black: (0, 0, 0)).
        thickness (int): The thickness of the circle outline. -1 for a filled circle.

    Returns:
        numpy.ndarray: The image with the circle drawn.
    """
    # Create a copy of the image to avoid modifying the original
    image_with_circle = image.copy()

    # Draw a circle on the image
    # cv2.circle(image, center_coordinates, radius, color, thickness)
    center_coordinates = (center_x, center_y)
    cv2.circle(image_with_circle, center_coordinates, radius, color, thickness)

    return image_with_circle

# Path to the user's nature image
image_path = '/content/nature image.jpeg'

# Read the image
input_image = cv2.imread(image_path)

# Check if image was loaded successfully
if input_image is None:
    print(f"Error: Could not load image from {image_path}")
else:
    # Define circle parameters
    circle_radius = 80
    circle_center_x = input_image.shape[1] // 2  # Center X-coordinate
    circle_center_y = input_image.shape[0] // 2  # Center Y-coordinate
    circle_color = (0, 0, 255) # Red color in BGR
    circle_thickness = 5 # Outline thickness

    # Draw the circle on the loaded image
    output_image = draw_circle_on_image(
        input_image, circle_radius, circle_center_x, circle_center_y,
        color=circle_color, thickness=circle_thickness
    )

    # Display the image using matplotlib
    # OpenCV uses BGR, Matplotlib uses RGB, so convert for correct display
    plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
    plt.title(f'Nature Image with a {circle_radius}-pixel Red Circle')
    plt.axis('off') # Hide axes
    plt.show()
```
