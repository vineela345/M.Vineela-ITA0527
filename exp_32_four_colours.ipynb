```python
from PIL import Image, ImageDraw

def create_image_with_corner_boxes(image_width, image_height):
    """
    Creates a white image of specified size with black, blue, green, and red
    boxes in each corner. Each box is 1/10th the size of the image.

    Args:
        image_width (int): The width of the image.
        image_height (int): The height of the image.

    Returns:
        PIL.Image.Image: The generated image.
    """
    # Create a new white image
    img = Image.new('RGB', (image_width, image_height), color = 'white')
    draw = ImageDraw.Draw(img)

    # Calculate box size (1/10th of image dimensions)
    box_width = image_width // 10
    box_height = image_height // 10

    # Define colors for the corners (RGB)
    black = (0, 0, 0)
    blue = (0, 0, 255)
    green = (0, 255, 0)
    red = (255, 0, 0)

    # Top-left corner (Black)
    draw.rectangle([(0, 0), (box_width, box_height)], fill=black)

    # Top-right corner (Blue)
    draw.rectangle([(image_width - box_width, 0), (image_width, box_height)], fill=blue)

    # Bottom-left corner (Green)
    draw.rectangle([(0, image_height - box_height), (box_width, image_height)], fill=green)

    # Bottom-right corner (Red)
    draw.rectangle([(image_width - box_width, image_height - box_height), (image_width, image_height)], fill=red)

    return img

# Example usage:
# Create an image of 500x300 pixels
image = create_image_with_corner_boxes(500, 300)

# Display the image
display(image)
```

```python
from PIL import Image, ImageDraw

def add_corner_boxes_to_image(image_path):
    """
    Loads an image from the given path and adds black, blue, green, and red
    boxes to each corner. Each box is 1/10th the size of the image.

    Args:
        image_path (str): The path to the input image file.

    Returns:
        PIL.Image.Image: The image with the added corner boxes.
    """
    try:
        img = Image.open(image_path).convert('RGB')
    except FileNotFoundError:
        print(f"Error: Image file not found at {image_path}")
        return None
    except Exception as e:
        print(f"Error loading image: {e}")
        return None

    draw = ImageDraw.Draw(img)
    image_width, image_height = img.size

    # Calculate box size (1/10th of image dimensions)
    box_width = image_width // 10
    box_height = image_height // 10

    # Define colors for the corners (RGB)
    black = (0, 0, 0)
    blue = (0, 0, 255)
    green = (0, 255, 0)
    red = (255, 0, 0)

    # Top-left corner (Black)
    draw.rectangle([(0, 0), (box_width, box_height)], fill=black)

    # Top-right corner (Blue)
    draw.rectangle([(image_width - box_width, 0), (image_width, box_height)], fill=blue)

    # Bottom-left corner (Green)
    draw.rectangle([(0, image_height - box_height), (box_width, image_height)], fill=green)

    # Bottom-right corner (Red)
    draw.rectangle([(image_width - box_width, image_height - box_height), (image_width, image_height)], fill=red)

    return img

# Path to your forest image
forest_image_path = '/content/forest image.avif'

# Add corner boxes to the forest image
modified_image = add_corner_boxes_to_image(forest_image_path)

# Display the modified image if it was successfully processed
if modified_image:
    display(modified_image)
```
