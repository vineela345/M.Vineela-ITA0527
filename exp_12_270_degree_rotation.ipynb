```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the user-provided image
original_image = Image.open('/content/forest image.jpg')

# 2. Perform a 270-degree clockwise rotation (equivalent to 90 degrees counter-clockwise)
# expand=True ensures the entire rotated image fits, expanding the canvas if necessary
rotated_image = original_image.rotate(90, expand=True)

# Display the original and rotated images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(original_image)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(rotated_image)
axes[1].set_title('Rotated Image (270° Clockwise)')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
