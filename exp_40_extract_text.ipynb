Now, here's a Python function that uses these libraries to extract text from a video. The function processes the video frame by frame, applies OCR, and collects the detected text.

```python
import cv2
import pytesseract
import os

def extract_text_from_video(video_path, lang='eng', frame_skip=10):
    """
    Extracts text from a video using OpenCV and Tesseract OCR.

    Args:
        video_path (str): The path to the video file.
        lang (str): The language to use for Tesseract OCR (e.g., 'eng' for English).
        frame_skip (int): Process every Nth frame to speed up extraction.

    Returns:
        list: A list of strings, where each string is text extracted from a frame.
    """
    all_extracted_text = []

    if not os.path.exists(video_path):
        print(f"Error: Video file not found at {video_path}")
        return []

    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print("Error: Could not open video file.")
        return []

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break  # End of video

        if frame_count % frame_skip == 0:
            # Convert frame to grayscale for better OCR performance
            gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

            # Use Tesseract to do OCR on the frame
            text = pytesseract.image_to_string(gray_frame, lang=lang)

            if text.strip(): # Only add non-empty text
                all_extracted_text.append(text.strip())

        frame_count += 1

    cap.release()
    print(f"Extracted text from {len(all_extracted_text)} frames.")
    return all_extracted_text

# Example usage (you'll need to upload a video file or provide a path to one):
# video_file_path = 'your_video.mp4' # Replace with your video file path
# extracted_texts = extract_text_from_video(video_file_path, frame_skip=30)
# for i, text in enumerate(extracted_texts):
#     print(f"--- Text from Frame {i * 30} ---")
#     print(text)
#     print("\n")
```

### Important Considerations:
*   **Performance:** Processing every frame can be very slow for long videos. The `frame_skip` parameter helps by processing only a subset of frames. Adjust it based on your video content and desired speed/accuracy balance.
*   **OCR Accuracy:** OCR quality depends heavily on video resolution, text clarity, font, lighting conditions, and the `lang` parameter. You might need to experiment with pre-processing steps (like resizing, sharpening, contrast adjustment) on the frames for better results.
*   **Duplicate Text:** This function will extract text from each selected frame. You might get a lot of duplicate text if the same text remains on screen for several frames. Post-processing to remove duplicates or aggregate text over time would be beneficial for many applications.
*   **Tesseract Languages:** To use other languages, you might need to install additional Tesseract language packs (e.g., `!sudo apt install tesseract-ocr-deu` for German) and specify the `lang` parameter accordingly (e.g., `lang='deu'`).
