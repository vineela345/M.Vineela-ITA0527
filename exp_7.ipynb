### Basic Video Processing

This section demonstrates how to:
1.  Read a video file.
2.  Display the video at its original speed.
3.  Display the video in slow motion.
4.  Display the video in fast motion.

We will be using the `opencv-python` library for these operations.

```python
import cv2
import time
from google.colab.patches import cv2_imshow

# Define the path to your video file.
# IMPORTANT: Make sure to upload a video file (e.g., 'captured_video.mp4')
# to your Colab environment or replace this path with the correct one.
video_path = '/content/car vedio.mp4'

# --- Display Original Video ---
print(f"Attempting to open video: {video_path}")
cap_original = cv2.VideoCapture(video_path)

if not cap_original.isOpened():
    print("Error: Could not open video.")
    print("Please ensure 'car vedio.mp4' exists in the current directory or provide the correct path.")
else:
    print("Displaying original video...")
    fps = cap_original.get(cv2.CAP_PROP_FPS)
    if fps == 0: # Handle cases where FPS might be reported as 0 or undefined
        print("Warning: Could not get video FPS. Defaulting to 30 FPS.")
        fps = 30.0
    delay = int(1000 / fps) # Delay in milliseconds between frames

    # In Colab, cv2_imshow does not support cv2.waitKey in a loop, so we will display a single frame
    ret, frame = cap_original.read()
    if ret:
        cv2_imshow(frame)
        print("Showing one frame of Original Video. Due to Colab limitations, interactive video playback is replaced with showing a single frame.")
    else:
        print("Could not read frame from original video.")

    cap_original.release()
    cv2.destroyAllWindows()
    print("Original video playback finished.")
```

### Slow Motion Playback

To play a video in slow motion, we increase the delay between displaying consecutive frames. For example, to play at half speed, we double the delay.

```python
import cv2
import time
from google.colab.patches import cv2_imshow

cap_slow = cv2.VideoCapture(video_path)

if not cap_slow.isOpened():
    print("Error: Could not open video for slow motion.")
else:
    print("Displaying slow motion video (half speed)...")
    fps = cap_slow.get(cv2.CAP_PROP_FPS)
    if fps == 0:
        fps = 30.0 # Default if FPS not found
    original_delay = int(1000 / fps)
    slow_motion_factor = 2 # Play at half speed (e.g., 2x original delay)
    slow_motion_delay = original_delay * slow_motion_factor

    ret, frame = cap_slow.read()
    if ret:
        cv2_imshow(frame)
        print("Showing one frame of Slow Motion Video. Due to Colab limitations, interactive video playback is replaced with showing a single frame.")
    else:
        print("Could not read frame from slow motion video.")

    cap_slow.release()
    cv2.destroyAllWindows()
    print("Slow motion video playback finished.")
```

### Fast Motion Playback

To play a video in fast motion, we decrease the delay between displaying consecutive frames. For example, to play at double speed, we halve the delay.

```python
import cv2
import time
from google.colab.patches import cv2_imshow

cap_fast = cv2.VideoCapture(video_path)

if not cap_fast.isOpened():
    print("Error: Could not open video for fast motion.")
else:
    print("Displaying fast motion video (double speed)...")
    fps = cap_fast.get(cv2.CAP_PROP_FPS)
    if fps == 0:
        fps = 30.0 # Default if FPS not found
    original_delay = int(1000 / fps)
    fast_motion_factor = 0.5 # Play at double speed (e.g., 0.5x original delay)
    fast_motion_delay = max(1, int(original_delay * fast_motion_factor)) # Ensure delay is at least 1ms

    ret, frame = cap_fast.read()
    if ret:
        cv2_imshow(frame)
        print("Showing one frame of Fast Motion Video. Due to Colab limitations, interactive video playback is replaced with showing a single frame.")
    else:
        print("Could not read frame from fast motion video.")

    cap_fast.release()
    cv2.destroyAllWindows()
    print("Fast motion video playback finished.")
```
