```python
from PIL import Image

# Directly load the specified image file
image_path = '/content/parrot image.avif'
try:
    image = Image.open(image_path)

    # Display original image
    print("\nOriginal Image:")
    display(image)

    # Perform 90-degree clockwise rotation
    # In PIL, positive angles are counter-clockwise. So, -90 degrees is clockwise.
    # `expand=True` ensures the entire rotated image fits without cropping,
    # adjusting the canvas size if necessary.
    rotated_image = image.rotate(-90, expand=True)

    # Display rotated image
    print("\nImage rotated 90 degrees clockwise:")
    display(rotated_image)
except FileNotFoundError:
    print(f"Error: The file '{image_path}' was not found. Please ensure it is in the '/content/' directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```
