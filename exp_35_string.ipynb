```python
import cv2
import numpy as np

def add_text_to_image(image_path, text, org, font_face, font_scale, color, thickness=2):
    """
    Adds text to an image and displays it.

    Args:
        image_path (str): Path to the input image. If None, a blank image will be created.
        text (str): The text string to add to the image.
        org (tuple): The bottom-left corner of the text string in the image (x, y).
        font_face (int): Font type, e.g., cv2.FONT_HERSHEY_SIMPLEX.
        font_scale (float): Font scale factor multiplied by the font-specific base size.
        color (tuple): Text color (B, G, R).
        thickness (int, optional): Thickness of the lines used to draw the text. Defaults to 2.
    """

    if image_path:
        img = cv2.imread(image_path)
        if img is None:
            print(f"Error: Could not load image from {image_path}. Creating a blank image instead.")
            img = np.zeros((400, 600, 3), np.uint8) # Create a black image if loading fails
            img.fill(255) # Make it white
    else:
        # Create a blank white image for demonstration
        img = np.zeros((400, 600, 3), np.uint8)
        img.fill(255) # Make it white

    # Add the text to the image
    cv2.putText(img, text, org, font_face, font_scale, color, thickness, cv2.LINE_AA)

    # Display the image
    # For Colab, cv2_imshow is preferred over cv2.imshow
    try:
        from google.colab.patches import cv2_imshow
        cv2_imshow(img)
    except ImportError:
        print("cv2_imshow not found. Using cv2.imshow (might not work in all environments).")
        cv2.imshow('Image with Text', img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()


# --- Example Usage ---
# Define text properties
user_text = "Nature's Beauty!"
position = (50, 100) # (x, y) coordinates for the bottom-left corner of the text
font = cv2.FONT_HERSHEY_DUPLEX
scale = 1.0
text_color = (0, 255, 0) # BGR color format (Green)
thickness = 2

# Call the function with the 'nature image.jpeg' file
add_text_to_image(image_path='/content/nature image.jpeg', text=user_text, org=position,
                  font_face=font, font_scale=scale, color=text_color, thickness=thickness)


# Example with a hypothetical image file (uncomment to test if you have an image)
# If you have an image file (e.g., 'my_image.jpg') in your Colab environment,
# you can uncomment and modify the line below:
# add_text_to_image(image_path='my_image.jpg', text='My Custom Text', org=(10, 50),
#                   font_face=cv2.FONT_HERSHEY_DUPLEX, font_scale=1.0, color=(255, 0, 0), thickness=2)
```
